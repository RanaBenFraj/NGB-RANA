```python
top_10_duration = data.nlargest(10, 'duration_ms')
print(top_10_duration)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(top_10_duration['track_name'], top_10_duration['duration_ms'])
```

```python
plt.scatter(data['loudness'], data['track_pop'], color='blue')
```

```python
plt.hist(data['danceability'], label='Danceability', color='blue')
plt.hist(data['energy'], label='Energy', color='orange')
```

```python
avg_danceability = data.groupby('album')['danceability'].mean().sort_values()
top_20_albums = avg_danceability.head(20)
top_20_albums.plot(kind='barh')
```
